```python
from fuzzywuzzy import fuzz

# Read files
df_a = pd.read_csv('a.csv')
df_b = pd.read_csv('b.csv')

# Define threshold for matching
threshold = 90

# Initialize empty list to store matching rows
matching_rows = []

# Loop through each row in file a
for i, row_a in df_a.iterrows():
    # Loop through each row in file b
    for j, row_b in df_b.iterrows():
        # Calculate Levenshtein distance between the 'Brewery' columns
        dist = fuzz.ratio(row_a['Brewery'], row_b['Brewery'])
        # If the distance is above the threshold, add the matching rows to the list
        if dist >= threshold:
            matching_rows.append((i, j))

# Create a new dataframe with the matching rows
df_matching = pd.DataFrame(matching_rows, columns=['index_a', 'index_b'])
df_matching = df_matching.merge(df_a, left_on='index_a', right_index=True)
df_matching = df_matching.merge(df_b, left_on='index_b', right_index=True)
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Load data from csv files
df_a = pd.read_csv('a.csv')
df_b = pd.read_csv('b.csv')

# Merge the Beer Name (Full) columns from both dataframes into one
df_a_b = pd.concat([df_a['Brewery'], df_b['Brewery']])
df_a_b = df_a_b.reset_index(drop=True)

# Create a TfidfVectorizer object to convert the beer names into vectors
vectorizer = TfidfVectorizer()

# Fit the vectorizer on the combined beer names
vectors = vectorizer.fit_transform(df_a_b)

# Calculate the cosine similarity matrix
cosine_sim_matrix = cosine_similarity(vectors)

# Create an empty list to store the matches
matches = []

# Iterate over each row in df_a and find the closest match in df_b using cosine similarity
for index, row_a in df_a.iterrows():
    beer_a = row_a['Brewery']
    vector_a = vectorizer.transform([beer_a])
    best_score = 0
    best_match = ''
    for index, row_b in df_b.iterrows():
        beer_b = row_b['Brewery']
        vector_b = vectorizer.transform([beer_b])
        score = cosine_similarity(vector_a, vector_b)[0][0]
        if score > best_score:
            best_score = score
            best_match = beer_b
    matches.append((beer_a, best_match, best_score))

# Create a dataframe from the matches list
df_matches = pd.DataFrame(matches, columns=['Brewery A', 'Brewery B', 'Score'])

# Filter the rows where the score is greater than or equal to 0.8
df_filtered_matches = df_matches[df_matches['Score'] >= 0.8]

# Print the filtered matches
print(df_filtered_matches)
```
